```python
import numpy as np
import time

# 9x9 초기문제 생성, 빈칸은 0으로 입력

init_array = [[0 for i in range(9)]
              for j in range(9)]

for i in range(9):
    print(init_array[i])

for i in range(9):
    for j in range(9):
        init_array[i][j] = input()



start = time.time()

number = 48

while number > 1:

    # init_array 출력 

    #for i in range(9):
    #    print(init_array[i])

    cand_array = [[0 for i in range(9)]
                  for j in range(9)]

    # 스도쿠 내에서 미지수와 결정된 수에 대한 새로운 후보군 배열 생성

    for i in range(9):
        for j in range(9):
            if init_array[i][j] == 0:
                cand_array[i][j] = [1,2,3,4,5,6,7,8,9]
    #            print(cand_array)
            else:
                cand_array[i][j] = int(init_array[i][j])

    #for i in range(9):
    #    print(init_array[i])

    #print('-----')

    #for i in range(9):
    #    print(cand_array[i])
    #print(cand_array[0][0])

    # init_array의 정수화 및 numpy 배열로 변경

    for i in range(9):
        for j in range(9):
            init_array[i][j] = int(init_array[i][j])

    init_array = np.array(init_array)

    #print(init_array)
    #print('-----')

    #for i in range(9):
    #    print(cand_array[i])

    # init_array의 행, 열 조사 및 cand_array의 후보군에서의 제거

    for i in range(9):
        for j in range(9):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[:,j])
                del_arr.append(init_array[i,:])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                for k in range(len(del_arr)):
                    cand_array[i][j].remove(del_arr[k])

            else:

                cand_array[i][j] = init_array[i][j]

    #print('------')

    #for i in range(9):
    #    print(cand_array[i])

    # 첫번째 박스

    for i in range(3):
        for j in range(3):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[0:3, 0:3])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # 두번째 박스

    for i in range(3):
        for j in range(3,6):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[0:3, 3:6])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]    

    # 세번째 박스

    for i in range(3):
        for j in range(6,9):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[0:3, 6:9])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]  

    # 네번째 박스

    for i in range(3,6):
        for j in range(3):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[3:6, 0:3])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # 다섯번째 박스

    for i in range(3,6):
        for j in range(3,6):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[3:6, 3:6])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # 여섯번째 박스

    for i in range(3,6):
        for j in range(6,9):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[3:6, 6:9])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # 일곱번째 박스

    for i in range(6,9):
        for j in range(3):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[6:9, 0:3])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # 여덟번째 박스

    for i in range(6,9):
        for j in range(3,6):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[6:9, 3:6])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # 아홉번째 박스

    for i in range(6,9):
        for j in range(6,9):

            if init_array[i][j] == 0:

                del_arr = []
                del_arr.append(init_array[6:9, 6:9])

                del_arr = np.reshape(del_arr, -1) # 1차원 배열로 변경
                del_arr = np.delete(del_arr, np.where(del_arr == 0)) # 0제거
                del_arr = np.unique(del_arr) # 중복 수 제거

                cand_array[i][j] = [x for x in cand_array[i][j] if x not in del_arr]

            else:

                cand_array[i][j] = init_array[i][j]

    # cand_array에서 원소의 개수가 1인 원소 추출, final_array에 할당

    final_array = [[0 for i in range(9)] for j in range(9)]


    for i in range(9):
        for j in range(9):  

            if type(cand_array[i][j]) == list and len(cand_array[i][j]) == 1:
                final_array[i][j] = cand_array[i][j]

            elif type(cand_array[i][j]) != list:
                final_array[i][j] = cand_array[i][j]

    num = 0

#    for i in range(9):
#        print(final_array[i])  

    for i in range(9):
        for j in range(9):
            if final_array[i][j] == 0:
                num = num + 1

    print('남은 미지수 개수:', num)
    
    number = num

    # 전체 코드의 반복을 위해 final_array의 정수화 및 init_array에 재할당

    for i in range(9):
        for j in range(9):
            if type(final_array[i][j]) == list:
                final_array[i][j] = final_array[i][j][0]

    for i in range(9):
        for j in range(9):
            init_array[i][j] = final_array[i][j]

#    for i in range(9):
#        print(init_array[i])



print('Solve!')

print('Solution is :\n', init_array)
print('time for solve:', time.time() - start)
```
